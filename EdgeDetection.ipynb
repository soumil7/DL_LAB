```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image_path = "thalaivar.jpg"  
original_image = cv2.imread(image_path, 0)

edges = cv2.Canny(original_image, 100, 150) 


plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Edges")
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```
