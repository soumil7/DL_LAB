```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "thalaivar.jpg"  
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
kernel = np.ones((5, 5))
```

```python
eroded_image = cv2.erode(original_image, kernel, iterations=1)


dilated_image = cv2.dilate(original_image, kernel, iterations=1)

# cv2.MORPH_ERODE: Erosion
# cv2.MORPH_DILATE: Dilation
# cv2.MORPH_OPEN: Opening (erosion followed by dilation)
# cv2.MORPH_CLOSE: Closing (dilation followed by erosion)
# cv2.MORPH_GRADIENT: Morphological gradient (difference between dilation and erosion)
# cv2.MORPH_TOPHAT: Top hat operation
# cv2.MORPH_BLACKHAT: Black hat operation
```

```python
open_img = cv2.morphologyEx(original_image, cv2.MORPH_OPEN, kernel, iterations=1)

close_img = cv2.morphologyEx(original_image, cv2.MORPH_CLOSE, kernel, iterations=1)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title("Dilated Image")
plt.imshow(dilated_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title("Open Image")
plt.imshow(open_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title("Close Image")
plt.imshow(close_img, cmap='gray')
plt.axis('off')
```
