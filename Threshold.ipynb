```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'thalaivar.jpg' 
original_image = cv2.imread(img_path, 0)
```

```python
threshold_value = 127
```

```python
_, binary_image = cv2.threshold(original_image, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
_, binaryinv_image = cv2.threshold(original_image, threshold_value, 255, cv2.THRESH_BINARY_INV)
```

```python
_, trunc_image = cv2.threshold(original_image, threshold_value, 255, cv2.THRESH_TRUNC)
```

```python
_, tozero_image = cv2.threshold(original_image, threshold_value, 255, cv2.THRESH_TOZERO)
```

```python
_, tozeroinv_image = cv2.threshold(original_image, threshold_value, 255, cv2.THRESH_TOZERO_INV)
```

```python
plt.figure(figsize=(8, 4))
```

```python
plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title("BINARY Image")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title("BINARY_INV Image")
plt.imshow(binaryinv_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title("TRUNC Image")
plt.imshow(trunc_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title("TOZERO Image")
plt.imshow(tozero_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.title("TOZERO_INV Image")
plt.imshow(tozero_image, cmap='gray')
plt.axis('off')
```
